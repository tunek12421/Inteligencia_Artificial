```python
# ==============================================
# EJERCICIO 3 (F√ÅCIL): Regresi√≥n lineal simple con datos sint√©ticos
# Dataset: Generado manualmente
# T√©cnica: Regresi√≥n lineal con una sola variable (una caracter√≠stica)
# ==============================================
# Paso 1: Importar bibliotecas necesarias
from sklearn.linear_model import LinearRegression # Modelo de regresi√≥n lineal


import matplotlib.pyplot as plt
import numpy as np
# Paso 2: Crear datos de ejemplo (poblaci√≥n vs ganancia)
poblacion = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 6.0, 6.5]) # en millones
ganancia = np.array([1.0, 1.5, 2.0, 2.5, 2.7, 3.5, 3.8, 4.2, 5.0, 5.5]) # en cientos de miles de d√≥lares
# Paso 3: Redimensionar datos (para que sklearn los acepte)
X = poblacion.reshape(-1, 1) # X debe ser una matriz 2D
y = ganancia # y puede ser un vector 1D
# Paso 4: Crear el modelo de regresi√≥n lineal
model = LinearRegression()
model.fit(X, y) # Entrenamos el modelo con los datos
# Paso 5: Visualizar los datos y la recta de regresi√≥n
plt.scatter(poblacion, ganancia, color='blue', label='Datos reales')
plt.plot(poblacion, model.predict(X), color='red', label='L√≠nea de regresi√≥n')
plt.title("Regresi√≥n lineal: Poblaci√≥n vs Ganancia")
plt.xlabel("Poblaci√≥n (millones)")
plt.ylabel("Ganancia (cientos de miles de $)")
plt.legend()
plt.grid(True)
#plt.show()
#plt.savefig('confusion_matrix_regresionLinear.png', dpi=300, bbox_inches='tight')
#print("Matriz de confusi√≥n guardada como 'confusion_matrix_regresionLinear.png'")
# Paso 6: Interpretar los coeficientes
print(f"Pendiente (coeficiente): {model.coef_[0]:.2f}")
print(f"Intersecci√≥n (ordenada al origen): {model.intercept_:.2f}")
# Explicaci√≥n:
# - La pendiente indica cu√°nto cambia la ganancia por cada mill√≥n adicional de personas.
# - La intersecci√≥n representa la ganancia estimada si la poblaci√≥n fuera 0.
# - La regresi√≥n lineal intenta ajustar la mejor l√≠nea recta que minimiza el error cuadr√°tico entre los puntos reales y la l√≠nea predicha.
```
